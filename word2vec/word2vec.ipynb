```python
import string
from gensim.models import Word2Vec
import PyPDF2
```

# Minecraft

```python
with open('minecraft.txt', 'r', encoding='utf-8') as file:
    documento = file.read()
```

```python
len(documento)
```

```python
oraciones = documento.split('.')
len(oraciones)
```

```python
oraciones[125]
```

```python
oraciones_limpias = []
for oracion in oraciones:
    tokens = oracion.translate(str.maketrans('','', string.punctuation)).split()
    tokens = [word.lower() for word in tokens if word.isalpha()]
    if tokens:
        oraciones_limpias.append(tokens)
```

```python
model = Word2Vec(sentences=oraciones_limpias, vector_size=500, 
                 window=5, min_count=1, workers=4)
```

```python
vector = model.wv['minecraft']
vector
```

```python
palabras_cercanas = model.wv.most_similar('construcción', topn=10)
palabras_cercanas
```

```python
model.save('minecraft.model')
modelo_cargado = Word2Vec.load('minecraft.model')
```

```python
model.wv.save_word2vec_format('minecraft_model.txt')
# model.wv.save_word2vec_format('minecraft_model.bin', binary=True)
# En binario ocupa mucho menos espacio, se usa más
```

```python
from gensim.models import KeyedVectors

embeddings_cargados = KeyedVectors.load_word2vec_format('minecraft_model.txt', 
                                                        binary=False)
embeddings_cargados
```

```python
def analogias(v1,v2,v3):
    similitud = embeddings_cargados.most_similar(positive=[v1,v3], negative=[v2])
    print(f'{v1} es a {v2} como {similitud[0][0]} es a {v3}')
```

```python
analogias('minecraft', 'videojuego', 'piedra')
```

# Frankenstein

```python
def extraer_texto_pdf(ruta_archivo):
    with open(ruta_archivo, '+rb') as archivo:
        lector = PyPDF2.PdfReader(archivo)
        texto = ''
        for pagina in range(len(lector.pages)):
            texto += lector.pages[pagina].extract_text()
    return texto

documento = extraer_texto_pdf('D:\\GitHub\LLM_1\\word2vec\\frankenstein.pdf')
```

# Modelo con más archivos

```python
import os 
from tqdm import tqdm
```

```python
# Usamos la función que creamos al principio de "Frankenstein", extraer_texto_pdf

ruta_carpeta = 'textos'
todos_los_textos = []

for archivo in tqdm(os.listdir(ruta_carpeta)):
    ruta_completa = os.path.join(ruta_carpeta, archivo)
    try:
        documento = extraer_texto_pdf(ruta_completa)
        todos_los_textos.append(documento)
    except Exception as e:
        print(f'Error al procesar {archivo}: {e}')
```

```python
oraciones_totales = []
tamaño = 0

for documento in todos_los_textos:
    tamaño = tamaño + len(documento)
    oraciones = documento.split('.')
    oraciones_totales.extend(oraciones)
tamaño
```

```python
len(oraciones_totales)
```

```python
oraciones_limpias_big = []
for oracion in oraciones_totales:
    tokens = oracion.translate(str.maketrans('','', string.punctuation)).split()
    tokens = [word.lower() for word in tokens if word.isalpha()]
    if tokens:
        oraciones_limpias_big.append(tokens)
```

```python
oraciones_limpias_big[46366]
```

```python
model_big = Word2Vec(sentences=oraciones_limpias_big, vector_size=1000, 
                 window=5, min_count=1, workers=5)
```

```python
palabras_cercanas = model_big.wv.most_similar('libertad', topn=10)
palabras_cercanas
```

```python
model_big.wv.save_word2vec_format('model_100.bin', binary=True)
```

```python
model_big_cargado = KeyedVectors.load_word2vec_format('model_100.bin', binary=True)
def analogias(v1,v2,v3):
    similitud = model_big_cargado.most_similar(positive=[v1,v3], negative=[v2])
    print(f'{v1} es a {v2} como {similitud[0][0]} es a {v3}')
```
